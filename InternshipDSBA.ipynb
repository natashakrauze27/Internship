```python
import copy
import random
'''
This part is dedicated to implementation of Karger's algorithm
It takes graph as a dict, where keys are vertecies and values are 
lists of incedent vertecies and returns list of verticies of the first
set of out cut
'''
def choose_random_key(G):
    v1 = random.choice(list(G.keys()))
    v2 = random.choice(G[v1])
    return v1, v2

def karger(G):
    length = []
    while len(G) > 2:
        v1, v2 = choose_random_key(G)
        G[v1].extend(G[v2])
        for x in G[v2]:
            G[x].remove(v2)
            G[x].append(v1)
        while v1 in G[v1]:
            G[v1].remove(v1)
        del G[v2]
    for key in G.keys():
        length.append(len(G[key]))
    return length[0]

def operation(n):
    i = 0
    count = 10000
    while i < n:
        data = copy.deepcopy(G)
        min_cut = karger(data)
        if min_cut < count:
            count = min_cut
        i = i + 1
    return min_cut

cuts=[]
def kargerMinCut(graph):
    while len(graph) > 2:
        v = random.choice(list(graph.keys()))
        w = random.choice(graph[v])
        contract(graph, v, w)
    mincut = len(graph[list(graph.keys())[0]])
    cuts.append(mincut)
def contract(graph, v, w):
    for node in graph[w]:
        if node != v:
            graph[v].append(node)
        graph[node].remove(w) 
        if node != v:
            graph[node].append(v)
    del graph[w]
```
